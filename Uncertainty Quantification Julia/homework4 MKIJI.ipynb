## Problem 1

One class of Orthogonal Polynomials we have not discussed in class are Chebyshev Polynomials

$$ \begin{align}
T_0(x) &= 1 \\
T_1(x) &= x \\
T_{n+1}(x) &= 2xT_n(x) - T_{n-1}(x),
\end{align}$$

which are orthogonal with respect to the weight

$$ w(x) = \begin{cases} 
      \dfrac{1}{\sqrt{1-x^2}} & -1\leq x\leq 1 \\
      0 & \text{else} 
   \end{cases} .$$

Show that the $T_1(x)$ and $T_2(x)$ polynomials are orthogonal.

$T_1(x)=x, T_2(x)=2x^2-1$

So $<T_1(x),T_2(x)> $ $=\int^1_{-1}T_1(x)*T_2(x)*w(x)dx \\ = \int^1_{-1}\frac{(2x^3-x)}{\sqrt{1-x^2}}dx \\ $

$
\text{Now}\, \int\frac{(2x^3-x)}{\sqrt{1-x^2}}dx \\=
\frac{1}{2}\int\frac{(2u+1)}{\sqrt{u}}du \, \text{where}\, u=x^2-1 \\
= \int(2\sqrt{u}+\frac{1}{\sqrt{u}})du \\
= 2\int\sqrt{u}\,du + \int\frac{1}{\sqrt{u}}\,du \\
= \frac{4u^{\frac{3}{2}}}{3} + 2\sqrt{u} + C\\
= \frac{2(x^2-1)^{\frac{3}{2}}}{3} + \sqrt{x^2-1} + C\\
= \frac{\sqrt{x^2-1}(2x^2+1)}{3}+C$

Therefore,

$\int^1_{-1}\frac{(2x^3-x)}{\sqrt{1-x^2}}dx \\ 
=\frac{\sqrt{1^2-1}(2*1^2+1)}{3}-(\frac{\sqrt{(-1^2)-1}(2*(-1^2)+1)}{3})
=0$

So $T_1(x)$ and $T_2(x)$ polynomials are orthogonal

## Problem 3

Given the model

$$ f(x,q) = q x + \sin(q) x^2, $$

where $q$ is distributed as an Exponential with $\lambda=1$.  Use Laguerre orthogonal polynomials, defined as

$$ \begin{align}
    L_0(x) &= 1 \\
    L_1(x) &= 1-x \\
    L_{n+1}(x) &= \left(\frac{2k+1-x}{k+1}\right)L_n(x) - \left(\frac{k}{k+1}\right)L_{n-1}(x),
    \end{align}
$$

and orthogonal with respect to the weight function

$$ w(x) = \begin{cases} 
      e^{-x} & x\geq0 \\
      0 & \text{else} 
   \end{cases} ,$$
   
to find the **second order** spectral approximation of $f(x,q)$ using the discrete projection method:

$$ f(x,q) \approx f_0(x) L_0(q) + f_1(x) L_1(q) + f_2(x) L_2(q).$$

Note: If the integrals become too challenging, feel free to use a computer algebra system.

$L_0(x) = 1, L_1(x) = 1-x, L_2(x)= \frac{1}{2}(x^2-4x+2) \\
\text{So}\, f(x,q) = q x + \sin(q) x^2 \approx f_0(x)+f_1(x)(1-q)+f_2(x)*\frac{1}{2}(q^2-4q+2)$

Now, to find $f_0(x)$, take inner product with $L_0(q)$

$<qx+\sin(q)x^2, 1> = <f_0(x)+f_1(x)(1-q)+f_2(x)*\frac{1}{2}(q^2-4q+2), 1> = f_0(x)<1,1> \\ 
\Leftrightarrow f_0(x)=\frac{<qx+\sin(q)x^2,1>}{<1,1>} \because L_0(q), L_1(q),\,\text{and}\, L_2(q)\, \text{are orthogoal}\,$

Then

$<qx+\sin(q)x^2,1>\, = \,\int^\infty_0\{qx+\sin(q)x^2\}e^{-q}\,dq = \frac{x^2+2x}{2}$ 

and 

$<1,1>\, = \,\int^\infty_0e^{-q}dq = 1$

Therefore, 

$f_0(x) = \frac{x^2+2x}{2}$

To find $f_1(x)$, take inner product with $L_1(q)$

$<qx+\sin(q)x^2, 1-q>\\ =

$<qx+\sin(q)x^2,1-q>\, = \,\int^\infty_0\{qx+\sin(q)x^2\}(1-q)e^{-q}\,dq = -x$

$<1-q,1-q>\,=\,\int^\infty_0(1-q)^2e^{-q}\,dq = 1$

Therefore, 

$f_1(x) = -x$

## Problem 4

Use Julia to plot both your true model (with random samples) and approximation for $x\in [-10,10]$.  Compare the mean and standard deviations of the true model and the approximation.

```julia
using Distributions, Random, Statistics, Plots; gr();
```

```julia
qdist = Normal(0,1)
```

```julia
f(x,q) = q*x + sin(q)*x^2
```

```julia
g(x,q) = (x^2+2*x)/2 - x*(1-q) -x^2*(q^2-4*q+2)/8
```

```julia
N=200
tval = -10:0.1:10
fvals = zeros(length(tval),N)

for i=1:N
    qval = rand(qdist)
    fvals[:,i] = f.(tval,qval)
end
```

```julia
plot(tval, fvals,color=:black, label="" )
plot!(tval, mean(fvals,dims=2), color=:red, label="Mean", linewidth=6)
```

```julia
gvals = zeros(length(tval), N)

for i=1:N
    qval = rand(qdist)
    gvals[:,i] = g.(tval,qval)
end
```

```julia
plot(tval, mean(fvals, dims=2), label="True Mean", linewidth=3, color=:red)
plot!(tval, mean(fvals, dims=2) +std(fvals,dims=2), label="True STD", linewidth=3, color=:red, linestyle=:dash)
plot!(tval, mean(fvals, dims=2) - std(fvals, dims=2), label="True STD", linewidth=3, color=:red, linestyle=:dash)
```

```julia
plot!(tval, mean(gvals, dims=2), label="SE Mean", linewidth=3, color=:green)
plot!(tval, mean(gvals, dims=2) + std(gvals,dims=2), label="SE STD", linewidth=3, color=:green, linestyle=:dash)
plot!(tval, mean(gvals, dims=2) - std(gvals, dims=2), label="SE STD", linewidth=3, color=:green, linestyle=:dash)
```
