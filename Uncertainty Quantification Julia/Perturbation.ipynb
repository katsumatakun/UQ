```julia
using Distributions, Random, Statistics, Plots; gr();
```

```julia
f(x,q) = q*x^2 + x*(q-1)^2
```

```julia
# Assuming q = 1 + eq
f_approx(x,eq) = x^2 + eq*x^2 + x*eq^2
```

```julia
xvec = -5:0.1:5
plot(xvec,f.(xvec,1))
```

```julia
N = 10000
fsamples = zeros(length(xvec),N)

for i = 1:N
    qval = randn()*0.1 + 1.0
    fsamples[:,i] = f.(xvec,qval)
end
```

```julia
plot(xvec,fsamples[:,1:100:end],label="",color=:black,alpha=0.2,
    xlabel="X (not q)",size=(600,400))
```

```julia
plot!(xvec,mean(fsamples,dims=2),label="Sample Mean", 
    color=:blue,linewidth=3)
plot!(xvec,mean(fsamples,dims=2) + 2 .* std(fsamples,dims=2),
    label="Sample 2*Std", color=:red,linewidth=3)
plot!(xvec,mean(fsamples,dims=2) - 2 .* std(fsamples,dims=2),
    label="", color=:red,linewidth=3)
```

```julia
plot!(xvec,f_approx.(xvec,0),color=:lightgreen,
    linewidth=3,linestyle=:dash,label="Perturbation Mean")
plot!(xvec,f_approx.(xvec,0.2),color=:red,
    linewidth=3,linestyle=:dash,label="Perturbation 2*Std")
plot!(xvec,f_approx.(xvec,-0.2),color=:red,
    linewidth=3,linestyle=:dash,label="")
```

```julia
N = 10000
fsamples_fixed = zeros(N)
qval = zeros(N)
xval = 2.0

for i = 1:N
    qval[i] = randn()*0.1 + 1.0
    fsamples_fixed[i] = f.(xval,qval[i])
end
```

```julia
histogram(fsamples_fixed,normed=true,xlabel="f(2,q)",label="Samples")
```

```julia
vline!([mean(fsamples_fixed)],color=:blue,linewidth=5,
    label="Sample Mean")
vline!([mean(fsamples_fixed) + 2 .* std(fsamples_fixed)],
    color=:red,linewidth=5,label="Sample 2*Std")
vline!([mean(fsamples_fixed) - 2 .* std(fsamples_fixed)],
    color=:red,linewidth=5,label="")
```

```julia
vline!([f_approx.(xval,0)],color=:lightgreen,linewidth=5,
    linestyle=:dash,label="Perturbation Mean")
vline!([f_approx.(xval,0.2)],color=:red,linewidth=5,
    linestyle=:dash,label="Perturbation 2*Std")
vline!([f_approx.(xval,-0.2)],color=:red,linewidth=5,
    linestyle=:dash,label="")
```

```julia
using Distributions, Random, Statistics, Plots; gr();
```

```julia
g(x,q1,q2) = (q1^2)*x^2 + x*(q2-1)^2
```

```julia
g_approx(x,eq1,eq2) = x^2 + (2*x^2*eq1 + 0) + (0 + 2*x*eq2^2)
```

```julia
xvec = -5:0.1:5
plot(xvec,g.(xvec,1,1))
```

```julia
N = 10000
gsamples = zeros(length(xvec),N)

for i = 1:N
    qval1 = randn()*0.1 + 1.0
    qval2 = randn()*0.1 + 1.0
    gsamples[:,i] = g.(xvec,qval1,qval2)
end
```

```julia
plot(xvec,gsamples[:,1:100:end],label="",
    color=:black,alpha=0.2,xlabel="X (not q)",size=(600,400))
```

```julia
plot!(xvec,mean(gsamples,dims=2),label="Sample Mean", 
    color=:blue,linewidth=3)
plot!(xvec,mean(gsamples,dims=2) + 2 .* std(gsamples,dims=2),
    label="Sample 2*Std", color=:red,linewidth=3)
plot!(xvec,mean(gsamples,dims=2) - 2 .* std(gsamples,dims=2),
    label="", color=:red,linewidth=3)
```

```julia
plot!(xvec,g_approx.(xvec,0,0),color=:lightgreen,
    linewidth=3,linestyle=:dash,label="Perturbation Mean")
plot!(xvec,g_approx.(xvec,0.2,0.2),color=:red,
    linewidth=3,linestyle=:dash,label="Perturbation 2*Std")
plot!(xvec,g_approx.(xvec,-0.2,-0.2),color=:red,
    linewidth=3,linestyle=:dash,label="")
```

```julia
N = 10000
gsamples_fixed = zeros(N)
qval1 = zeros(N)
qval2 = zeros(N)
xval = 2.0

for i = 1:N
    qval1[i] = randn()*0.1 + 1.0
    qval2[i] = randn()*0.1 + 1.0
    gsamples_fixed[i] = g.(xval,qval1[i],qval2[i])
end
```

```julia
histogram(gsamples_fixed,normed=true,xlabel="f(2,q)",label="Samples")
```

```julia
vline!([mean(gsamples_fixed)],color=:blue,linewidth=5,label="Sample Mean")
vline!([mean(gsamples_fixed) + 2 .* std(gsamples_fixed)],
    color=:red,linewidth=5,label="Sample 2*Std")
vline!([mean(gsamples_fixed) - 2 .* std(gsamples_fixed)],
    color=:red,linewidth=5,label="")
```
