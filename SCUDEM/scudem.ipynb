```julia
using DifferentialEquations, ParameterizedFunctions, Plots; gr();
```

```julia
plotly();
```

```julia
X = @ode_def begin
    db = -x*b +10*b*(L-b)/L-k_1*w*10*b*(L-b)/L
    dw = k_2*10*b*(L-b)/L*(R-w)-5*x*w
    end x L R k_1 k_2 
```

```julia
R_S = [400.0,100.0]
tspan = (0.0,0.05)
params = [0.3,1000,200,0.25, 0.5]
prob = ODEProblem(X,R_S,tspan, params)
```

```julia
sol = solve(prob)
sol.(tspan[2])
plot(sol)
```
