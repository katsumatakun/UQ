```julia
N = 1000

x_dist = Normal(0,10)
x = rand(x_dist, N)

y_dist = Normal(10,2.5)
y = rand(y_dist, N)
```

```julia
scatter(x,y)
```

```julia
histogram(x)
```

```julia
histogram(y)
```

```julia
A = [ones(N) x]
A_t = transpose(A)
b = y
x = inv(A_t*A)*A_t*b
```

```julia
N = 1000

x_dist = Uniform(0,25)
x = rand(x_dist, N)

y_dist = Normal(0,2.5)
y = 3*x .+ 30 .+ rand(y_dist, N)
```

```julia
plt = plot(layout=(1,2), size=(400,200), legend=false)
scatter!(plt[1],x,y)
histogram!(plt[2],x)
```

```julia
A = [ones(N) x]
A_t = transpose(A)
b = y
x = inv(A_t*A)*A_t*b
```

```julia
N = 1000

x_dist = Uniform(0,25)
x = rand(x_dist, N)

y_dist = Normal(0,50)
y = 30 .+ 10*x - 5*x.^2 + rand(y_dist, N)
```

```julia
plt = plot(layout=(1,3), size=(500,200), legend=false)
scatter!(plt[1],x,y)
histogram!(plt[2],x)
histogram!(plt[3],y)
```

```julia
A = [ones(N) x]
A_t = transpose(A)
b = y
beta = inv(A_t*A)*A_t*b
```

```julia
scatter!(plt[1],x,beta[1] .+ beta[2]*x, color=:red)
```

```julia
A = [ones(N) x x.^2 x.^3 x.^4 x.^5]
A_t = transpose(A)
b = y
beta = inv(A_t*A)*A_t*b
```

```julia
scatter!(plt[1],x,beta[1] .+ beta[2]*x .+beta[3]*x.^2, color=:green)
```

```julia
N = 1000

x_dist = Uniform(0,25)
x = rand(x_dist, N)

y_dist = Normal(0,1)
y = 5 .+ 5*sin.(x) + rand(y_dist, N)
```

```julia
scatter(x,y)
```

```julia
A = [ones(N) sin.(x)]
# A = [ones(N) x x.^2 x.^3]
A_t = transpose(A)
b = y
beta = inv(A_t*A)*A_t*b
```

```julia
scatter!(x,beta[1].+beta[2]*sin.(x), color=:pink)
# scatter!(x,beta[1].+beta[2]*x + beta[3]*x.^2 + beta[4]*x.^3, color=:red)
```

```julia
yhat = beta[1].+beta[2]*sin.(x)
```

```julia
sqrt(sum((y-yhat).^2))
```
