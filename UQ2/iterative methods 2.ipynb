# Iterative Methods 2
## Kaczmarz Algorithm

```julia
using Plots, LinearAlgebra
```

```julia
function myKaczmarz(x,A,b,i)
    x = x + ((b[i] - A[i,:]' * x)/(A[i,:]' * A[i,:])) * A[i,:]
    
    return x, mod(i,2)+1
end
```

```julia
x = -2:0.1:3
y1 = ones(length(x))
y2 = x .- 1
# mod(3,3)
```

```julia
plot(x,y1,linewidth=4)
plot!(x,y2,linewidth=4)
```

```julia
p1 = [0; 0]
i = 1;
```

```julia
## <0,1> is a direction vector perpendicular to the line 0*x+1*y=k
## <-1,1> is a direcrion vector perpendicular to the line -1*x+1*y=k
A = [0 1; -1 1]
```

```julia
b = [1; -1]
```

```julia
A*[2; 1]
```

```julia
p1c = p1
p2, i = myKaczmarz(p1,A,b,i)
p1 = p2
plot!([p1c[1],p2[1]],[p1c[2],p2[2]],marker=:circle,color=:black)
```
